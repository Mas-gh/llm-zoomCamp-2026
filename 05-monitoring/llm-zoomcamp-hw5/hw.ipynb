```python
from opentelemetry import trace
from opentelemetry.sdk.trace import TracerProvider
from opentelemetry.sdk.trace.export import ConsoleSpanExporter, SimpleSpanProcessor

provider = TracerProvider()
provider.add_span_processor(SimpleSpanProcessor(ConsoleSpanExporter()))
trace.set_tracer_provider(provider)
tracer = trace.get_tracer("llm-zoomcamp")

# Cell: Defining the Traced RAG
from starter import rag # This imports your RAGBase

class RAGTraced:
    def __init__(self, rag_instance):
        self.rag = rag_instance
    
    def search(self, query):
        with tracer.start_as_current_span("search") as span:
            results = self.rag.search(query)
            return results
            
    def llm(self, prompt):
        with tracer.start_as_current_span("llm") as span:
            response = self.rag.llm(prompt)
            # You will add token attributes here for Q2
            return response
            
    def rag(self, query):
        with tracer.start_as_current_span("rag") as span:
            return self.rag.rag(query)

# Initialize the traced version
traced_rag = RAGTraced(rag)
```
